```python
date_column = 'Unnamed: 1'
tonality_column = 'Unnamed: 5'
sm_index_column = 'Unnamed: 6'
```

```python
print(df['Unnamed: 1'].head(10))
```

```python
df[date_column] = pd.to_datetime(df[date_column], format='%d.%m.%Y %H:%M').dt.date
```

```python
df[sm_index_column] = df[sm_index_column].fillna(0)
```

```python
def plot_monthly_data(df, month, year):
    
    monthly_data = df[(pd.DatetimeIndex(df[date_column]).month == month) & (pd.DatetimeIndex(df[date_column]).year == year)]
    
    if monthly_data.empty:
        print(f'Нет данных за {month}/{year}')
        return
    
    
    daily_data = monthly_data.groupby(date_column).agg(
        messages=(date_column, 'size'),
        sm_index=(sm_index_column, 'sum'),
        positive=(tonality_column, lambda x: (x == 'позитивная').sum()),
        neutral=(tonality_column, lambda x: (x == 'нейтральная').sum()),
        negative=(tonality_column, lambda x: (x == 'негативная').sum())
    ).reset_index()

    fig, axs = plt.subplots(3, 1, figsize=(12, 18))
    
    # График динамики сообщений
    axs[0].plot(daily_data[date_column], daily_data['messages'], marker='o')
    axs[0].set_title('Динамика сообщений')
    axs[0].set_xlabel('Дата')
    axs[0].set_ylabel('Количество сообщений')
    axs[0].grid(True)

    axs[1].plot(daily_data[date_column], daily_data['sm_index'], marker='o', color='orange')
    axs[1].set_title('Динамика СМ Индекса')
    axs[1].set_xlabel('Дата')
    axs[1].set_ylabel('СМ Индекс')
    axs[1].grid(True)

    width = 0.2
    dates = np.arange(len(daily_data[date_column]))
    axs[2].bar(dates - width, daily_data['positive'], width, label='Позитивная')
    axs[2].bar(dates, daily_data['neutral'], width, label='Нейтральная')
    axs[2].bar(dates + width, daily_data['negative'], width, label='Негативная')
    axs[2].set_title('Динамика тональности')
    axs[2].set_xlabel('Дата')
    axs[2].set_ylabel('Количество сообщений')
    axs[2].set_xticks(dates)
    axs[2].set_xticklabels(daily_data[date_column], rotation=45)
    axs[2].legend()
    axs[2].grid(True)
    
    plt.tight_layout()
    plt.savefig(f'monthly_data_{year}_{month}.png')
    plt.show()
    
    daily_data.to_excel(f'monthly_data_{year}_{month}.xlsx', index=False)
    print(f'Данные за {month}/{year} обработаны и сохранены.')

for month in [9, 10, 11, 12]:
    plot_monthly_data(df, month, 2023)
```
